```python
import numpy as np
```

```python
data = np.array([
    [25, 5000, 4.5, 1, 0],
    [45, 8000, 12.2, 0, 1],
    [35, 6000, 5, 1, 1],
])
data
```

```python
#Scaling just the first three elements in each row

data[:, :3] = (data[:, :3] - data[:, :3].mean()) / data[:, :3].std() # standarding by subtracting the mean and dividing by the standard deviation
data
```
