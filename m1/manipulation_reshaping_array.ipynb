#### We can join (concatenate) multiple NumPy arrays

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
arr3 = np.array([7, 8, 9])
```

```python
arr4 = np.concatenate((arr1, arr2, arr3))
```

```python
arr4
```

```python
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6]])
```

```python
mat3 = np.concatenate((mat1,mat2))
```

```python
mat3
```

```python
mat4 = np.array([[7], [8]])
```

```python
mat5 = np.concatenate((mat1, mat4), axis = 1)
```
