## Indexing and Slicing Arrays - 1

```python
my_arr = np.array([1, 2, 3, 4, 5])
```

```python
my_arr[0]
```

```python
my_arr[-1]
```

### slicing

```python
my_arr[2:] # starts slicing from 2nd index to last 
```

```python
my_arr[:2] # starts slicing from 0th index to 2nd index
```

```python
my_arr[::2] # start and end value will take as 0 and step is 2 here 
```

#### We can use boolean indexing to retrieve only those values we mark as True. We can mark values in this way by using conditions

```python
ages_list = [1, 18, 20, 23, 24, 25, 27, 8]
```

```python
# ageages_list >=18 # cannot compare lists to integers 
```

```python
ages = np.array([ages_list])
```

```python
ages>=18
```

```python
ages[ages>=18]
```

#### We can use 'fancy indexing' to select only certain items by index from a NumPy array

```python
my_arr[[3, 0, 1]] # my_arr = array([1, 2, 3, 4, 5])
```

```python
my_arr[[3, 1, 1, 1]]
```

#### We can use np.where() to get the indices where the elements within a NumPy array meet a condition

```python
indices = np.where(my_arr % 2 == 0)
```

```python
indices # (array([1, 3]),)
```

```python
my_arr[indices] # array([2, 4]), value of the index
```

#### We can use np.where() to modify NumPy arrays

```python
np.where(my_arr > 2, 100, my_arr) # ( condition, case if met , else); does not act in place 
```

```python
new_arr = np.where(my_arr < 4, my_arr ** 2, -55)
```

```python
new_arr # array([  1,   4,   9, -55, -55])
```

#### We can also directly modify NumPy arrays using subscripts

```python
my_arr_2 = my_arr # shallow copy 
```

```python
my_arr_2
```

```python
my_arr_2[my_arr_2 > 2 ] = 100
```

```python
my_arr_2
```

```python
my_arr # array([  1,   2, 100, 100, 100])
```

```python
my_arr = np.array([1, 2, 3, 4, 5]) # restoring my_arr to its initial values
```

```python
my_arr
```

```python
my_arr_3 = my_arr.copy()
```

```python
my_arr_3
```

```python
my_arr_3[my_arr_3 > 2 ] = 100
```

```python
my_arr_3 # array([  1,   2, 100, 100, 100])
```

```python
my_arr # array([1, 2, 3, 4, 5])
```

#### We can slice multidimensional array by using commas in subscripts

```python
matrix = np.array([[[1, 2, 3], [6, 7, 8]], [[1, 2, 2], [6, 7, 7]]])
```

```python
matrix
```

```python
matrix[0][1][2] # np.int64(8)
```

```python
my_mat = matrix[matrix > 2]
```

```python
my_mat
```

```python
matrix[::1]
# array([[[1, 2, 3],
#         [6, 7, 8]],

#        [[1, 2, 2],
#         [6, 7, 7]]])
```

```python
matrix[::2]
# array([[[1, 2, 3],
        # [6, 7, 8]]])
```
