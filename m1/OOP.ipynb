```python
#Define a class to represnt a customer

class Customer:

    # Constructor to initialize customer details 

    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.purchase_history = []

    # Method to add a purchase 
    def add_purchase(self, item):
        self.purchase_history.append(item)

    # Method to display customer info 
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Email: {self.email}")
        print(f"Purchases: {self.purchase_history}")
```

```python
# Create an object (instance) of the Customer class

customer1 = Customer("Viplav", "viplav.soni@bounteous.com")
```

```python
customer1.add_purchase("Laptop")
customer1.add_purchase("Earphones")
```

```python
# Display all customer details

customer1.display_info()
```

## The __init__() Constructor

```python
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Emp Id: {self.emp_id}")
        print(f"Department: {self.department}")
```

```python
emp1 = Employee("Viplav Soni", 123, "Ops Tech")
emp2 = Employee("Alok", 23, "PB Tech")
```

```python
print(emp1.display_info())
print("-------------------")
print(emp2.display_info())
```

## Class Variables & Class Methods

- **Class Variables**: 
While instance variables are specific to each object, class variables are shared across all instances of the class. These are typically used to store data that should remain consistent for all objects. They are defined outside of any method, directly under the class definition.

- **Class Methods**:
To interact with class variables or perform operations that affect the class as a whole, we use class methods. These are marked with the @classmethod decorator and take cls as the first argument instead of self.

```python
# Define an Employee class with a class variable 

class Employee:

    totalEmployee = 0 

    def __init__(self, name, department):
        self.name = name
        self.department = department
        Employee.totalEmployee += 1

    def display_info(self):
        print(f"Name: {self.name}, Department: {self.department}")

    @classmethod
    def get_taotal_employee(cls):
        return cls.totalEmployee
```

```python
emp1 = Employee("Viplav Soni", "PB Tech")
emp2 = Employee("Prashant", "IT")
```

```python
emp1.display_info()
emp2.display_info()
```

```python
print("Total Employees:", Employee.get_taotal_employee())
```

## Static Methods (@staticmetod)

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_info(self):
        print(f"Name: {self.name}, Salary: {self.salary}")

    @staticmethod
    def calculate_tax(salary):
        return salary * 0.8
```

```python
emp1 = Employee("Viplav Soni", 80000)
emp2 = Employee("Prashant", 120000)
```

```python
emp1.display_info()
emp2.display_info()
```

```python
# call the static method to calculate tax

tax = Employee.calculate_tax(emp1.salary)
print(f"Estimated Income Tax: {tax}")
```

```python
tax = Employee.calculate_tax(emp2.salary)
print(f"Estimated Income Tax: {tax}")
```
