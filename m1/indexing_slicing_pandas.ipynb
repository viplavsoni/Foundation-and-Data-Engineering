Sync two devices data

```python
phone_steps = pd.Series(data = [1000, 4000, 7400, 8000, 2200], index = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], name = 'Phone')
phone_steps
```

```python
watch_steps = pd.Series(data = [1000, 7400,2200], index = ['Mon', 'Wed', 'Fri'], name = 'Watch')
```

```python
total_steps = phone_steps + watch_steps
```

```python
total_steps
```

```python
phone_steps.add(watch_steps, fill_value= 0 ) # set the missing values to 0 before adding
```

```python
phone_steps.add(watch_steps, fill_value= 100 )
```

We can fill missing values with `.fillna()`

```python
total_steps.fillna(9999)
```

```python
total_steps
```

```python
total_steps.isna()
```

We can use `inplace` parameter to perform an operation without having to reassign the Series to a variable; most methods on Series can use this parameter 

```python
tmp = total_steps.fillna(12345, inplace=True)
```

```python
type(tmp)
```

```python
total_steps
```

Drop missing values with `.dropna()`

```python
total_steps = phone_steps + watch_steps
```

```python
total_steps
```

```python
total_steps.dropna()
```

```python
total_steps
```

```python
total_steps.dropna(inplace=True)
```

```python
total_steps
```

```python
total_steps.rank()
```

```python
total_steps.reset_index()
```
