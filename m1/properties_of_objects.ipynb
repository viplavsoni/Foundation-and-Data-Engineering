### Inheritance in Python

**Why Use Inheritance?**

- Avoid duplication: Write common logic once in a base class
- Organise related classes: Create clear relationships between objects
- Extend functionality: Add or modify features in subclasses without changing the original code

#### Business Use Case: Employee and Managers

```python
# Base Class: Employee

class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_info(self):
        print(f"Name: {self.name}, Salary: {self.salary}")
```

```python
# Child Class: Manager ( Inherits from Employee )
```

```python
class Manager(Employee):

    def __init__(self, name, salary, team_size):
        # Call the parent class constructor
        super().__init__(name, salary)
        self.team_size = team_size

    # Add additional method for manager
    def display_manager_info(self):
        self.display_info() # Call method from parent
        print(f"Manages a team of {self.team_size} people")
```

```python
m1 =  Manager("Viplav", 40000, 10)
```

```python
m1.display_manager_info()
```

### Method Overriding

Method overriding means defining a method in the child class with the same name as a method in the parent class, so that the child class version is used instead.


We need method overriding

 - to change or extend how inherited methods behave
 - to ensure subclass behaviour aligns with specific needs, while still using the general structure of the parent class


```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_details(self):
        print(f"Employee Name: {self.name}, Salary: {self.salary}")
```

```python
class Manager(Employee): 
    def __init__(self, name, salary, team_size):
        super().__init__(name, salary)
        self.team_size = team_size

    def get_details(self):
        print(f"Manager Name: {self.name}, Salary: {self.salary}, Team Size: {self.team_size}")
```

```python
emp1 = Employee("Viplav Soni", 20000)
manager1 = Manager("Ram", 60000, 10)
```

```python
emp1.get_details()
```

```python
manager1.get_details()
```

```python
Employee.get_details(manager1)
```

### Polymorphism

**polymorphism allows objects of different classes to respond to the same method call in a way that’s appropriate to their type.**

- It allows for writing general-purpose code that works across different object types
- Makes the system extensible, so new functionality can be added with minimal changes to the existing code
- Helps decouple logic from specific implementations

```python
class Sedan:
    def role(self):
        print('Nagin')

class Suv:    
    def role(self):
        print('Hooo Haaa')

#Ploymorphic behaviour
def show_horn(horn):
        horn.role()
```

```python
c1 = Sedan()
c2 = Suv()

show_horn(c1)
show_horn(c2)
```

### Abstraction

- Practice of hiding internal implementation details and showing only the essential features of an object or system.

- To create a blueprint or contract for future classes
- To hide complex logic from users who only need to interact with the high-level interface
- To enforce consistency across different class implementations

```python
from abc import ABC, abstractmethod

class Engineer(ABC):

    @abstractmethod
    def role(self):
        pass
```

```python
class Developer(Engineer):
    def role(self):
        print('Develop the code')

class Tester(Engineer):
    def role(self):
        print('Test the code')

engineers = [Developer(), Tester()]

for engineer in engineers:
    engineer.role()
```
