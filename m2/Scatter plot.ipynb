```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
df.info()
```

```python
plt.scatter(data = df, x = 'age', y = 'cholestrol')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(data = df, x = 'age', y = 'cholestrol')
plt.title('Plot of age versus cholestrol levels')
plt.xlabel('Cholestrol')
plt.ylabel('Age')
plt.show()
```

## Seaborn

```python
sns.scatterplot(data = df, x = 'age', y = 'cholestrol')
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', s = 10) # change the size of the markers
plt.title('Plot of age versus cholestrol levels')
plt.xlabel('Cholestrol')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', c = 'r', s= 10) # change the color of the markers
plt.title('Plot of age versus cholestrol levels')
plt.xlabel('Cholestrol')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', c ='g',alpha = 0.5) # change the opacity of the markers
plt.title('Plot of age versus cholestrol levels')
plt.xlabel('Cholestrol')
plt.ylabel('Age')
plt.show()
```

```python
plt.style.use('ggplot') # changing the plot style
plt.figure(figsize=(5, 3))
sns.scatterplot(data = df, x = 'age', y = 'cholestrol', c ='g',alpha = 0.5) # change the opacity of the markers
plt.title('Plot of age versus cholestrol levels')
plt.xlabel('Cholestrol')
plt.ylabel('Age')
plt.show()
```
