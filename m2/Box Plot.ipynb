```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('churn_data.csv')
```

```python
plt.figure(figsize = (5, 3))
sns.boxplot(data = df, y = 'MonthlyCharges')
plt.title('Boxplot of monthly charges')
plt.ylabel('Monthly charges')
plt.show()
```

We could break up the data by payment type:

```python
plt.figure(figsize = (5, 3))
sns.boxplot(
    data = df,
    x = 'PaymentMethod',
    y = 'MonthlyCharges',
    order = sorted(list(df['PaymentMethod'].unique()))
)
plt.title('Boxplot of monthly charges for each payment method')
plt.xlabel('Payment method')
plt.xticks(rotation = 20)
plt.ylabel('Monthly charges')
plt.show()
```
