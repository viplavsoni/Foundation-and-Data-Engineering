```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AirPassengers.csv')
```

```python
df
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 3))
sns.lineplot(data=df, x = 'Month', y ='Passengers' )
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()
```

```python
ax = sns.lineplot(data=df, x = 'Month', y ='Passengers' )
x_tick_labels = ax.get_xticklabels()
plt.close()
```

```python
x_tick_labels
```

```python
x_tick_labels_array = np.array(x_tick_labels)
x_tick_labels_array[range(23, 144, 24)]
```

```python
plt.figure(figsize=(8, 3))
sns.lineplot(data=df, x = 'Month', y ='Passengers' )
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))
plt.ylabel('Passengers')
plt.show()
```

```python
plt.figure(figsize = (8, 3))
sns.lineplot(
    data = df,
    x = 'Month',
    y = 'Passengers',
    marker = 'o',
    markersize = 4,
    markerfacecolor = 'orange',
    lw = 1,
    ls = ':'
)
plt.title('Line plot of monthly number of passengers in the airline')
plt.xlabel('Month')
plt.xticks(range(23, 144, 24))
plt.ylabel('Passengers')
plt.ylim((0, 700))  # setting a range for the y-axis limits
plt.show()
```
