```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
sns.histplot(data = df, x= 'age')
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(data = df, x= 'age', bins=5)
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(data = df, x= 'age', bins=range(20, 80, 10))
plt.xlabel('Age')
plt.show()
```

Let's draw different histogram based on sex

```python
df_male = df [df['sex'] == 1]
df_female = df [df['sex'] == 0]
```

```python
df_male
```

```python
sns.histplot(data = df_male, x= 'age')
sns.histplot(data = df_female, x= 'age')
plt.title('Distribution of ages of people based on sex')
plt.xlabel('Age')
plt.show()
```
