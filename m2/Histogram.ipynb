```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
df
```

```python
sns.histplot(data = df, x= 'age')
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(data = df, x= 'age', bins=5)
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(data = df, x= 'age', bins=range(20, 80, 10))
plt.xlabel('Age')
plt.show()
```

Let's draw different histogram based on sex

```python
df_male = df [df['sex'] == 1]
df_female = df [df['sex'] == 0]
```

```python
df_male
```

```python
sns.histplot(data = df_male, x= 'age', label = 'male') # label will show in legend
sns.histplot(data = df_female, x= 'age', label = 'female')
plt.title('Distribution of ages of people based on sex')
plt.xlabel('Age')
plt.legend() # add a legend
plt.show()
```

```python
sns.histplot(data = df, x= 'age', hue='sex') # label will show in legend

plt.title('Distribution of ages of people based on sex')
plt.xlabel('Age')

plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.histplot(data = df, x = 'age', hue = 'sex')
plt.title('Distribution of ages of people based on sex')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.show()
```

We could also present this data side-by-side as subplots within the same figure. Try to understand what the following code is doing before executing it:

```python
plt.figure(figsize = (8, 3))
ax1 = plt.subplot(1, 2, 1)
sns.histplot(data = df_male, x = 'age', color = 'blue')
plt.title('Histogram of ages of men')
plt.xlabel('Age')
plt.xticks(rotation = 45)
plt.ylabel('Number of people')

ax2 = plt.subplot(1, 2, 2, sharey = ax1)
sns.histplot(data = df_female, x = 'age', color = 'orange')
plt.title('Histogram of ages of women')
plt.xlabel('Age')
plt.xticks(rotation = 45)
plt.ylabel('')
ax2.tick_params(labelleft = False, length = 0)

plt.suptitle('Histograms of ages of men and women')
plt.tight_layout()
plt.show()
```
