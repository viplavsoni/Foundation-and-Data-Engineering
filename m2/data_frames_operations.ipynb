```python
import pandas as pd
import numpy as np
```

```python
data1 = {
    'Name': ['Ashok', 'Bob', 'Chandni', 'Dawood', 'Esha'],
    'Age': [25, 30, 18, 22, 30],
    'City': ['Mumbai', 'Kolkata', 'Delhi', 'Bengaluru', 'Agra'],
    'Inc June': [3000, 4000, 5000, 6000, np.nan],
    'Inc July': [3000, 4000, 5000, np.nan, 7000]
}
d1 = pd.DataFrame(data1)
```

```python
d1
```

```python
d1['Inc of June and July'] = d1['Inc June'] + d1['Inc July']
```

```python
d1
```

```python
d1['Inc of June and July'] = d1['Inc June'].add(d1['Inc July'], fill_value = 0) # .add() is a series method which works here
```

```python
d1
```

```python
d1['Tax on Income'] = 0.01 * d1['Inc of June and July']
```

```python
d1
```

```python
d1[['Inc June', 'Inc July']] = d1[['Inc June', 'Inc July']].fillna(4)
```

```python
d1
```

```python
d1.loc[d1['City'].isin(['Mumbai', 'Delhi']), 'Tax on Income'] +=500
```
