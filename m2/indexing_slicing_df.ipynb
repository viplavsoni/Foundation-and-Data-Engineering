We can index and slice DataFrames by position with iloc[], the integer-location-based (position-based) indexer

```python
df.iloc[0]
```

```python
df.iloc[::2]
```

```python
df.iloc[0, :]
```

```python
df.iloc[:2, :5]
```

```python
type(df.iloc[:2, :5]) # a selection of multiple rows and columns is 2D and thereforew a Pandas DataFrame
```

```python
df.iloc[::-1, ::-1] # rows and columns in reverse
```

```python
df.loc[54, 'symboling'] # here, the label of the row is the same as its numeric index (0)
```

```python
df.loc[:, 'symboling': 'wheelbase']
```

```python
df.loc[df['carbody']=='sedan',:]
```
