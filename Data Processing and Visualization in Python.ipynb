```python

def convertToweightCategory(wght):
    if wght > 120:
        return 'overweight'
    elif wght >= 90:
        return 'normalweight'
    else:
        return 'underweight'
```

```python
nba['weightCategory'] = nba['Weight(kgs)'].apply(convertToweightCategory)
```

```python
# Apples Production in Metric Tonn / Hectare

years = [ 2010 , 2011 , 2012 , 2013 , 2014, 2015 ]
yield_apples = [ 0.895 , 0.91 , 0.919  , 0.926 , 0.929 , 0.931 ]
```

```python
plt.title('Apples Production (Metric Tonn) per Hectare')
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tonns) per hectare')
plt.show()
```

```python
flight_df = sns.load_dataset('flights')
```

```python
flight_df.head()
```

```python
flights_pivot = flight_df.pivot(index = 'month', columns='year', values='passengers')
```

```python
flights_pivot
```

```python
plt.title("No. of passengers")
sns.heatmap(flights_pivot, annot=True, fmt='d')
plt.show()
```
